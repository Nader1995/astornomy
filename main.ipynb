# Star Cluster Catalog Analysis with Pandas

- PPMXL ----------------------------------> star/source ID
  
- RAJ2000[deg] ---------------------------> Right Ascension in degrees
  
- DEJ2000[deg] ---------------------------> Declination in degrees
  
- pmRA[mas/yr] --------------------------> proper motion in RA
  
- pmDE[mas/yr] --------------------------> proper motion in De
  
- e_pmRA[mas/yr] -----------------------> error in proper motion RA
  
- e_pmDE[mas/yr] -----------------------> error in proper motion Dec
  
- Jmag[mag] -------------------------------> J-band infrared magnitude
  
- e_Jmag[mag] ----------------------------> error in J magnitude
  
- Kmag[mag] ------------------------------> K-band infrared magnitude
  
- e_Kmag[mag] ---------------------------> error in K magnitude

## Drop NaN
First of all, let's see how many NaN values we have among all elements and then drop them. 
    - first count the NaN

```python
display(df.sample(10))
display(df.isnull().sum())
```

It seems there is only one column **e_Kmag[mag]** containing actually NaN, and the rest of them contain empty cells. Let's replace empty cells by NaN and the drop them:

```python
df = df.replace(
 
    r'^\s*$',
    np.nan,
    regex=True
)
```

```python
display(df.sample(10, random_state=53))
display(df.isnull().sum())
```

Second, we drop all rows containing NaN:

```python
display(f"Number of rows before dropping null values: {len(df)}")
df = df.dropna()
display(f"Number of rows after dropping null values: {len(df)}")
```
