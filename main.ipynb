# Star Cluster Catalog Analysis with Pandas

- PPMXL ----------------------------------> star/source ID
  
- RAJ2000[deg] ---------------------------> Right Ascension in degrees
  
- DEJ2000[deg] ---------------------------> Declination in degrees
  
- pmRA[mas/yr] --------------------------> proper motion in RA
  
- pmDE[mas/yr] --------------------------> proper motion in De
  
- e_pmRA[mas/yr] -----------------------> error in proper motion RA
  
- e_pmDE[mas/yr] -----------------------> error in proper motion Dec
  
- Jmag[mag] -------------------------------> J-band infrared magnitude
  
- e_Jmag[mag] ----------------------------> error in J magnitude
  
- Kmag[mag] ------------------------------> K-band infrared magnitude
  
- e_Kmag[mag] ---------------------------> error in K magnitude

## Drop NaN
First of all, let's see how many NaN values we have among all elements and then drop them. 
    - first count the NaN

```python
display(df.sample(10))
display(df.isnull().sum())
```

It seems there is only one column **e_Kmag[mag]** containing actually NaN, and the rest of them contain empty cells. Let's replace empty cells by NaN and the drop them:

```python
df = df.replace(
 
    r'^\s*$',
    np.nan,
    regex=True
)
```

```python
display(df.sample(10, random_state=53))
display(df.isnull().sum())
```

Second, we drop all rows containing NaN:

```python
display(f"Number of rows before dropping null values: {len(df)}")
df_cleaned = df.dropna()
display(f"Number of rows after dropping null values: {len(df_cleaned)}")
```

Now we find the rows for which Jmag > 15.5, display their total number and then drop them:
But before we get into that, let's check if any element of column Jmag is anything other than number. To do that, we convert all column to number. If it is already a number, it won't change, otherwise, if it is a string or anything else, it will change to NaN:

```python
df_cleaned.loc[:, 'Jmag[mag]'] = pd.to_numeric(df_cleaned['Jmag[mag]'], errors='coerce')

display(df_cleaned['Jmag[mag]'].isna().sum())
```

The code above assures us that all element of column **Jmag** is a valid number, so:

```python
df_cleaned_Jmag_deleted = df_cleaned[df_cleaned['Jmag[mag]'] > 15.5]
display(f"Number of rows to be deleted based on Jmag[mag] > 15.5: {len(df_cleaned_Jmag_deleted)}")

df_cleaned_Jmag = df_cleaned[df_cleaned['Jmag[mag]'] <= 15.5]   
display(f"Number of rows after filtering Jmag[mag] <= 15.5: {len(df_cleaned_Jmag)}")

display(f'Is it equal to the initial number of rows? {len(df_cleaned_Jmag) + len(df_cleaned_Jmag_deleted)==len(df_cleaned)}')
```

Next we filter our dataset based on Kmag <= 15.4:

```python
df_cleaned_Jmag.loc[:, 'Kmag[mag]'] = pd.to_numeric(df_cleaned_Jmag['Kmag[mag]'], errors='coerce')

df_cleaned_Kmag_deleted = df_cleaned_Jmag[df_cleaned_Jmag['Kmag[mag]'] > 15.4]
display(f"Number of rows to be deleted based on Kmag[mag] > 15.4: {len(df_cleaned_Kmag_deleted)}")

df_cleaned_Kmag = df_cleaned_Jmag[df_cleaned_Jmag['Kmag[mag]'] <= 15.4]   
display(f"Number of rows after filtering Kmag[mag] <= 15.4: {len(df_cleaned_Kmag)}")

display(f'Is it equal to the initial number of rows? {len(df_cleaned_Kmag) + len(df_cleaned_Kmag_deleted)==len(df_cleaned_Jmag)}')
```

Now we demand the error in proper motion (in Radian, e_pmRA) to be restricted between 4 and 10: 

```python
df_cleaned_Kmag.loc[:, 'e_pmRA[mas/yr]'] = pd.to_numeric(df_cleaned_Kmag['e_pmRA[mas/yr]'], errors='coerce')

df_cleaned_e_pmRA_deleted = df_cleaned_Kmag[(df_cleaned_Kmag['e_pmRA[mas/yr]'] > 10) | (df_cleaned_Kmag['e_pmRA[mas/yr]'] < 4)]
display(f"Number of rows to be deleted based on e_pmRA[mas/yr] > 10 or e_pmRA[mas/yr] < 4: {len(df_cleaned_e_pmRA_deleted)}")

df_cleaned_e_pmRA = df_cleaned_Kmag[(df_cleaned_Kmag['e_pmRA[mas/yr]'] <= 10) & (df_cleaned_Kmag['e_pmRA[mas/yr]'] >= 4)]
display(f"Number of rows after filtering e_pmRA[mas/yr] <= 10 and e_pmRA[mas/yr] >= 4: {len(df_cleaned_e_pmRA)}")

display(f'Is it equal to the initial number of rows? {len(df_cleaned_e_pmRA) + len(df_cleaned_e_pmRA_deleted)==len(df_cleaned_Kmag)}')
```

Now this time, we demand the error in proper motion (in Degree, e_pmDE) to be restricted between 4 and 10:

```python
df_cleaned_e_pmRA.loc[:, 'e_pmDE[mas/yr]'] = pd.to_numeric(df_cleaned_e_pmRA['e_pmDE[mas/yr]'], errors='coerce')

df_processed_data_deleted = df_cleaned_e_pmRA[(df_cleaned_e_pmRA['e_pmDE[mas/yr]'] > 10) | (df_cleaned_e_pmRA['e_pmDE[mas/yr]'] < 4)]
display(f"Number of rows to be deleted based on e_pmDE[mas/yr] > 10 or e_pmDE[mas/yr] < 4: {len(df_processed_data_deleted)}")

df_processed_data = df_cleaned_e_pmRA[(df_cleaned_e_pmRA['e_pmDE[mas/yr]'] <= 10) & (df_cleaned_e_pmRA['e_pmDE[mas/yr]'] >= 4)]
display(f"Number of rows after filtering e_pmDE[mas/yr] <= 10 and e_pmDE[mas/yr] >= 4: {len(df_processed_data)}")

display(f'Is it equal to the initial number of rows? {len(df_processed_data) + len(df_processed_data_deleted)==len(df_cleaned_e_pmRA)}')
```

Now we add another column, χ-squared (chi-squared), based on the correct columns:
(We copy df_processed_data to processed_data to avoid **SettingWithCopyWarning**)


```python
processed_data = df_processed_data.copy()

RA_c = 52.5
Dec_c = 49.0
d_c = 172.4

mu_ac = 22.73
mu_dc = -26.51
e_ac = 0.17
e_dc = 0.17

s_ac0 = 0.7
s_dc0 = 0.7

s_ac1 = s_ac0/(d_c*4.74*10**(-3))
s_dc1 = s_dc0/(d_c*4.74*10**(-3))

processed_data.loc[:,'Chi_squared'] = (
        ((processed_data['pmRA[mas/yr]'] - mu_ac) ** 2) /
        ((processed_data['e_pmRA[mas/yr]'] ** 2) + e_ac ** 2 + s_ac1 ** 2) +
        ((processed_data['pmDE[mas/yr]'] - mu_dc) ** 2) /
        ((processed_data['e_pmDE[mas/yr]'] ** 2) + e_dc ** 2 + s_dc1 ** 2)
)
```

Now that we have created a new column, we have to filter χ-squared and eliminate χ-squared >= 6.17

```python

final_processed_data_deleted = processed_data[processed_data['Chi_squared'] >= 6.17]
display(f"Number of rows to be deleted based on Chi_squared >= 6.17: {len(final_processed_data_deleted)}")

final_processed_data = processed_data[processed_data['Chi_squared'] < 6.17]
display(f"Number of rows after filtering Chi_squared < 6.17: {len(final_processed_data)}")

display(f'Is it equal to the initial number of rows? {len(final_processed_data) + len(final_processed_data_deleted)==len(processed_data)}')
```

Just the last check to make sure our filter is working fine:

```python
display(final_processed_data.loc[
    (final_processed_data['Chi_squared'] >= 6.17) | 
    (final_processed_data['e_pmDE[mas/yr]'] > 10) | 
    (final_processed_data['e_pmDE[mas/yr]'] < 4)  |
    (final_processed_data['e_pmRA[mas/yr]'] > 10) | 
    (final_processed_data['e_pmRA[mas/yr]'] < 4)  |
    (df_cleaned_Jmag['Kmag[mag]'] > 15.4)         |
    (df_cleaned['Jmag[mag]'] > 15.5)
    ])
```
