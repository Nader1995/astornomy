# Star Cluster Catalog Analysis with Pandas

- PPMXL ----------------------------------> star/source ID
  
- RAJ2000[deg] ---------------------------> Right Ascension in degrees
  
- DEJ2000[deg] ---------------------------> Declination in degrees
  
- pmRA[mas/yr] --------------------------> proper motion in RA
  
- pmDE[mas/yr] --------------------------> proper motion in De
  
- e_pmRA[mas/yr] -----------------------> error in proper motion RA
  
- e_pmDE[mas/yr] -----------------------> error in proper motion Dec
  
- Jmag[mag] -------------------------------> J-band infrared magnitude
  
- e_Jmag[mag] ----------------------------> error in J magnitude
  
- Kmag[mag] ------------------------------> K-band infrared magnitude
  
- e_Kmag[mag] ---------------------------> error in K magnitude
